```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def lotka_volterra(t, initial_populations, prey_birth_rate, predation_rate, predator_death_rate, predator_reproduction_rate):

    prey, predator = initial_populations
    prey_vals = [prey]
    predator_vals = [predator]

    for time in t[1:]:
        dprey = prey_birth_rate * prey - predation_rate * prey * predator
        dpredator = predator_death_rate * prey * predator - predator_reproduction_rate * predator

        prey += dprey
        predator += dpredator

        prey = max(1, prey)  # Ensure population doesn't go extinct
        predator = max(1, predator)

        prey_vals.append(prey)
        predator_vals.append(predator)

    return np.array(prey_vals), np.array(predator_vals)
```

```python
prey_birth_rate = 0.1
predation_rate = 0.02
predator_death_rate = 0.01
predator_reproduction_rate = 0.1

t = np.linspace(0, 200, 1000)

initial_populations = [30, 10]

prey_population, predator_population = lotka_volterra(t, initial_populations, prey_birth_rate, predation_rate, predator_death_rate, predator_reproduction_rate)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, prey_population, label='Prey Population', color='blue')
plt.plot(t, predator_population, label='Predator Population', color='red')
plt.title('Predator-Prey Model - Lotka-Volterra')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.grid(True)
plt.show()
```
